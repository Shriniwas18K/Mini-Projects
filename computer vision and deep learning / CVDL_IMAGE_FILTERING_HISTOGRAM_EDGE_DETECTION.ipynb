### HISTOGRAM

```python
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import cv2
import numpy as np

def show_hist(img:np.matrix):
  '''Takes input grayscale image and plots its histogram'''
  H=np.histogram(img,bins=255)
  plt.bar(H[1][0:255],H[0])
  plt.show()
src=cv2.imread("H.jpeg")
src=cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
cv2_imshow(src)
show_hist(src)
src1=cv2.equalizeHist(src)
cv2_imshow(src1)
show_hist(src1)
```

### FILTERING

####### AVERAGE BLUR for all pixels look like neighbouring pixels

```python
src2=cv2.blur(src,(5,5))
cv2_imshow(src2)
```

######## GAUSSIAN BLUR for naturally blurred image

```python
src3=cv2.GaussianBlur(src,(5,5),0)
cv2_imshow(src3)
```

######## MEDIAN BLUR for salt and pepper noise

```python
src4=cv2.medianBlur(src,5)
cv2_imshow(src4)
```

######## CONVOLVING ANY MASK

```python
src5=cv2.filter2D(src,-1,np.ones((2,2),np.float32)/4)
cv2_imshow(src5)
```

## EDGE DETECTION

###### GRADIENT BASED

####### SOBEL

```python
src5=cv2.Sobel(src,cv2.CV_64F,1,0,ksize=1)
cv2_imshow(src5)
```

####### SCHARR

```python
src6=cv2.Scharr(src,cv2.CV_64F,1,0,delta=20)
cv2_imshow(src6)
```

#### GAUSSIAN BASED

######## LAPLACIAN

```python
src7=cv2.Laplacian(src,cv2.CV_64F)
cv2_imshow(src7)
```

####### CANNY

```python
src8=cv2.Canny(src,100,200)
cv2_imshow(src8)
```
